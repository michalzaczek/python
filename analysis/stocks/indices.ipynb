```python
import pandas as pd
import pandas_datareader as pdr
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', '^FCHI']
```

```python
df = pd.DataFrame()
```

```python
for i in tickers:
    df[i] = pdr.DataReader(i, 'yahoo', '1997-1-1')['Adj Close']
```

```python
df.head()
```

```python
(df / df.iloc[0] * 100).plot(figsize=(15,10))
```

```python
returns = ((df / df.shift(1)) - 1).mean() * 250
```
