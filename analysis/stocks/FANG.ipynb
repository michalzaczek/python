```python
import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
import numpy as np
```

```python
start = dt.datetime(2013,1,1)
end = dt.datetime(2019,9,19)
```

```python
tickers = ['FB', 'AAPL', 'NFLX', 'GOOGL']
```

```python
data = pd.DataFrame()
```

```python
for i in tickers:
    data[i] = web.DataReader(i, 'yahoo', start, end)['Adj Close']
```

```python
data.head()
```

```python
data.iloc[0]
```

```python
(data / data.iloc[0]).plot(figsize=(8,5))
```

```python
weights = np.array([.25, .25, .25, .25])
```

```python
returns = (data / data.shift(1)) - 1
```

```python
returns.head()
```

```python
annual_returns = returns.mean() * 250
```

```python
np.dot(annual_returns, weights)
```
