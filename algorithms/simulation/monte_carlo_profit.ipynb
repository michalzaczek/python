```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
revenues = 170
rev_std = 20
iterations = 1000
```

```python
rev = np.random.normal(revenues, rev_std, iterations)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(rev)
plt.show()
```

```python
COGS = rev * np.random.normal(.6, .1)
```

```python
plt.figure(figsize=(15, 6))
plt.axhline(y=COGS.mean(), color='red')
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
gross_profit = rev - COGS
```

```python
gross_profit.max()
```

```python
gross_profit.min()
```

```python
gross_profit.mean()
```

```python
gross_profit.std()
```

```python
plt.figure(figsize=(10, 6))
plt.axvline(x=gross_profit.mean(), color='red')
plt.hist(gross_profit, bins=20)
plt.show()
```
